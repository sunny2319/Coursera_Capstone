```python
import pandas as pd
import numpy as np
```

```python
tor_data='Toronto_postalcode_canada.csv'
tor_postcodes=pd.read_csv(tor_data).set_index('Postcode')
tor_postcodes.rename_axis('Postal Code',axis='index',inplace=True)
tor_postcodes
```

```python
tor_geo_csvdata='https://cocl.us/Geospatial_data'
!wget -q -O 'toronto_m.geospatial_data.csv' tor_geo_csvdata
geo_csvdata=pd.read_csv(tor_geo_csvdata).set_index('Postal Code')
geo_csvdata.head()
```

Combining Two DataFrames with Mapped PostCodes

```python
tor_neighbour=tor_postcodes.join(geo_csvdata)
tor_neighbour.head()
```

```python
tor_neighbour.to_csv('tor_neighbour')
```
